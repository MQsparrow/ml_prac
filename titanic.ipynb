```python
from urllib.request import urlretrieve
url = "https://github.com/MQsparrow/ml0616/raw/main/train.csv"
urlretrieve(url, "train.csv")
url = "https://github.com/MQsparrow/ml0616/raw/main/test.csv"
urlretrieve(url, "test.csv")
```

```python
import pandas as pd
train_df = pd.read_csv("train.csv", encoding="utf-8")
predict_df = pd.read_csv("test.csv", encoding="utf-8")
```

```python
predict_df
```

```python
datas = pd.concat([train_df, predict_df], axis=0, ignore_index=True)
datas = datas.drop(["PassengerId", "Survived"], axis=1)
datas
```

```python
# 資料預處理
# step1: 添補空值
# 1.1 check空值數
s = datas.isna().sum()
# s[s!=0]: filter操作
s[s!=0].sort_values(ascending=False)
```

```python
# 1.2 check每個欄位型態
# a. 數值: Age, SibSp, Parch, Ticket(#share), Fare
# b. 類別: Pclass, Name(mid), Sex, Cabin(first), Embarked
# 補空值(最可能值)
# 類別: 出現頻率最高數值
# 數值: ***平均值(X)*** 中位數(O) => 兩者不相互影響
# 需求 = 最可能的值
```

```python
# 處理ticket/name/cabin
def cabinchange(c):
    if pd.isna(c):
        return None
    else:
        return c[0]
datas["Cabin"] = datas["Cabin"].apply(cabinchange)
```

```python
counts = datas["Ticket"].value_counts()
def ticketchange(t):
    if pd.isna(t):
        return None
    else:
        return counts[t]
datas["Ticket"] = datas["Ticket"].apply(ticketchange)
```

```python
import re
# n = "Braund, Mr. Owen Harris"
# .+:任意字一次以上/ , (.+):且group起來/ \.:點
def namechange(n):
    return n.split(", ")[-1].split(". ")[0].strip()
datas["Name"] = datas["Name"].apply(namechange)
datas["Name"]
```

```python
# 類別: 最常出現
most = datas["Embarked"].value_counts().idxmax()
datas["Embarked"] = datas["Embarked"].fillna(most)
#check還缺多少
s = datas.isna().sum()
s[s!=0].sort_values(ascending=False)
datas
```

```python
# 數值: 中位數
med = datas.median().drop(["Pclass"])
datas = datas.fillna(med)
#check還缺多少
s = datas.isna().sum()
s[s!=0].sort_values(ascending=False)
```

```python
# One-Hot一定要做, 若為美觀可把過少案例去掉
count = datas["Name"].value_counts()
reserved = count[count > 50].index
def namefilter(mid):
    if mid in reserved:
        return mid
    else:
        return None
datas["Name"] = datas["Name"].apply(namefilter)
```

```python
# Pclass特別處: 雖為類別但有大小關係, 所以不做可以
# 做不做各有各好處, 這裡建議做
# sex可不做: 因為2值
pd.set_option('display.max_columns', 50)
datas = pd.get_dummies(datas)
datas = pd.get_dummies(datas, columns=["Pclass"])
datas
```

```python
datas["Family"] = datas["SibSp"] + datas["Parch"]
datas
```
